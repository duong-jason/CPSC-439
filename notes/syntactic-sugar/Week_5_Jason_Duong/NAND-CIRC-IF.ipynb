```python
# The Setup
# Reference: Book's Code - https://github.com/boazbk/tcscode

"""
counter = 0
def NAND(a, b):
    global counter
    counter += 1
    v = f't{counter}'
    print(f'{v} = NAND({a}, {b})')
    return v
    # return 0 if a == b == 1 else 1
"""

def NAND(a, b):
    return 1 - a * b

def NOT(a):
    return NAND(a, a)

def AND(a, b):
    t1 = NAND(a, b)
    return NOT(t1)

def OR(a, b):
    t1 = NOT(a)
    t2 = NOT(b)
    return NAND(t1, t2)
```

```python
# QUESTION #1: EXERCISE 4.4 Part 1 - "Proof by Code": de-sugar the conditional statement

def COND(cond, b, c):
    """return cond ? b : c"""
    notcond = NAND(cond, cond)
    t1 = NAND(c, notcond)      # check if condition = False
    t2 = NAND(b, cond)         # check if condition = True
    return NAND(t1, t2)

print(''.join('COND({}, {}, {}) = {}\n'.format(a, b, c, COND(a, b, c))
      for a in [0, 1] for b in [0, 1] for c in [0, 1]))
```

```python
# QUESTION #2 - LOOKUP1 Implementation

def LOOKUP1(a, b, c):
    """return c ? b : a"""  
    return COND(c, b, a)

print(''.join('LOOKUP1({}, {}, {}) = {}\n'.format(a, b, c, LOOKUP1(a, b, c))
      for a in [0, 1] for b in [0, 1] for c in [0, 1]))
```

```python
# QUESTION #3 - LOOKUP2 Implementation

def LOOKUP2(a, b, c, d, e, f):
    """ return a ? (e, f) == (0, 0) :
               b ? (e, f) == (0, 1) :
               c ? (e, f) == (1, 0) : d
    """
    t1 = LOOKUP1(c, d, f)
    t2 = LOOKUP1(a, b, f)
    return COND(e, t1, t2)

print(''.join('LOOKUP2({}, {}, {}, {}, {}, {}) = {}\n'.format(a, b, c, d, e, f, LOOKUP2(a, b, c, d, e, f))
      for a in [0, 1] for b in [0, 1] for c in [0, 1] for d in [0, 1] for e in [0, 1] for f in [0, 1]))
```

```python
# QUESTION #4 - Straight-Line Program w/ only NAND-STATEMENTS

def LOOKUP2(a, b, c, d, e, f):
    t1 = NAND(f, f)
    t2 = NAND(c, t1)
    t3 = NAND(d, f)
    t4 = NAND(t2, t3)
    t5 = NAND(f, f)
    t6 = NAND(a, t5)
    t7 = NAND(b, f)
    t8 = NAND(t6, t7)
    t9 = NAND(e, e)
    t10 = NAND(t8, t9)
    t11 = NAND(t4, e)
    return NAND(t10, t11)
    
print(''.join('LOOKUP2({}, {}, {}, {}, {}, {}) = {}\n'.format(a, b, c, d, e, f, LOOKUP2(a, b, c, d, e, f))
      for a in [0, 1] for b in [0, 1] for c in [0, 1] for d in [0, 1] for e in [0, 1] for f in [0, 1]))
```
